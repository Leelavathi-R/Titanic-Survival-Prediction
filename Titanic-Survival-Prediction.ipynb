## 1. Load the datasets

```python
import pandas as pd

train_df = pd.read_csv('./data/train.csv')
test_df = pd.read_csv('./data/test.csv')
display(train_df.head(10))
display(test_df.head())
print('train dataset: ',train_df.shape)
print('test dataset: ',test_df.shape)
```

```python
train_df['Survived'].value_counts()
```

```python
train_df.head()
```

## 1. Data Cleaning

### 1. Handle duplicate rows

```python
train_df = train_df.drop_duplicates()
train_df
```

### 1.2. Handle missing values

## There are some missing values in the columns:
    - Age, Embarked of train_df
    - Age, Fare of test_df

```python
train_df.info()
test_df.info()
```

### Filling the missing values of 'Age' with 'median' since its a numerical columns and median is less prone to outliers, also preserves the overall distribution.

```python
def fill_age(data):
    data['Age'].fillna(data['Age'].median(), inplace=True)
    return data
train_df = fill_age(train_df)
test_df = fill_age(test_df)
```

### There are only 2 rows in train_df with missing values in 'Embarked' 
    - So, filling it with the mode value.

```python
train_df[train_df['Embarked'].isnull()]
train_df.Embarked.fillna(train_df['Embarked'].mode()[0], inplace = True)
```

```python
train_df['Embarked'].mode()[0]
```

### There is one row of test_df with missng value in Fare
    - So, find the appropriate value from train_df based on Pclass, Age, Sex, SibSp, Parch, Embarked.
    - Selecting 7.7750, which is the reasonable value bases on the above criteria.

```python
display(test_df[test_df['Fare'].isnull()])
display(train_df[(train_df['Pclass'] == 3) & (train_df['Sex'] == 'male') & (train_df['Age']>=60)])
test_df.Fare.fillna(7.7750, inplace = True)
display(test_df[test_df['Fare'].isnull()])
```

```python
train_df.isnull().sum()
test_df.isnull().sum()
```

## 2. EDA 
- EDA has been performed already [My GitHub](https://github.com/Leelavathi-R/Titanic-Survival-EDA). Based on that, the major influencing factors are: 
    - Age, Sex, Fare, Pclass, and Embarked.
    - So, dropping the other columns

```python
def drop_columns(data):
    drop_cols = ['PassengerId','Name','Ticket','SibSp','Parch','Cabin']
    data = data.drop(drop_cols, axis = 1)
    return data
```

## 3. Feature Engineering
    - To improve the model performance, created two more new features 'AgeGroup', 'FamilySize' from 'Age', and 'SibSp' + 'Parch'.

### 3.1 New feature 'FamilySize'

```python
train_df['FamilySize'] = train_df['SibSp'] +train_df['Parch']
test_df['FamilySize'] = test_df['SibSp'] +test_df['Parch']

train_df = drop_columns(train_df)
display(train_df.head())

test_ids = test_df['PassengerId']
test_df = drop_columns(test_df)
display(test_df.head())
```

### 3.2 New feature 'AgeGroup'

```python
bins = [0, 12, 18, 35, 60, 100] 
labels = ['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']


train_df['AgeGroup'] = pd.cut(train_df['Age'], bins=bins, labels=labels, right=False)
test_df['AgeGroup'] = pd.cut(test_df['Age'], bins=bins, labels=labels, right=False)
```

```python
train_df.head()
```

## 4. Data Transformation
    - Label encoding for 'Sex', 'AgeGroup', and 'Embarked'

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
cols = ['Sex','AgeGroup','Embarked']
for col in cols:
    train_df[col] = label_encoder.fit_transform(train_df[col])
    test_df[col] = label_encoder.fit_transform(test_df[col])
    print(label_encoder.classes_)

display(train_df.head())
display(test_df.head())
```

```python
train_df
```

## 5. Building Model -  logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score, confusion_matrix

x = train_df.loc[:, train_df.columns != 'Survived']
y = train_df['Survived']

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 0.2, random_state = 42)
print(x_train.shape,y_train.shape)
print(x_val.shape, y_val.shape)

model = LogisticRegression(random_state = 0, max_iter = 2000)
model.fit(x_train, y_train)

y_val_pred = model.predict(x_val)

print('Validation Accuracy:', accuracy_score(y_val, y_val_pred))
print('Confusion matrix(Validation):\n', confusion_matrix(y_val, y_val_pred))
```

### Testing

```python
y_test_pred = model.predict(test_df)

sub_df = pd.DataFrame({'PassengerId':test_ids.values,
                       'Survived': y_test_pred
    
})

sub_df.to_csv('./submission.csv', index = False)
```
